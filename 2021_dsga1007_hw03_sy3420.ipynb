https://en.wikipedia.org/wiki/Beta_distribution


Your task is that for every pair of these 16 pairs of $(a, b)$:

1. (20 points) To generate a random vector obtained by the beta distribution $B(a, b)$ of size `n_size = 100_000`. You can use the following function: 
https://numpy.org/doc/stable/reference/random/generated/numpy.random.beta.html
    
2. (30 points) And for this random vector, to plot **as a subplot** the corresponding histogram with the number of bins `n_bins = 50`.



Important Note: your solution should produce only one figure with $4 \times 4$ subplots (i.e. 4 subplots in each row and 4 in each column).

Other Note: "100_000" is a convenience notation for "100000"
    

```python
#generate beta function:
a=0.5
b=1.2
c=2.0
d=5.0

x1=np.random.beta(a, b, size=100000)
x2=np.random.beta(a, c, size=100000)
x3=np.random.beta(a, d, size=100000)
x4=np.random.beta(a, a, size=100000)
x5=np.random.beta(b, a, size=100000)
x6=np.random.beta(b, b, size=100000)
x7=np.random.beta(b, c, size=100000)
x8=np.random.beta(b, d, size=100000)
x9=np.random.beta(c, a, size=100000)
x10=np.random.beta(c, b, size=100000)
x11=np.random.beta(c, c, size=100000)
x12=np.random.beta(c, d, size=100000)
x13=np.random.beta(d, a, size=100000)
x14=np.random.beta(d, b, size=100000)
x15=np.random.beta(d, c, size=100000)
x16=np.random.beta(d, d, size=100000)
```

```python
#plot
fig = plt.figure()
n_bin=50
fig, axes = plt.subplots(4, 4, sharey=True, tight_layout=True)

axes[0, 0].hist(x1, bins=n_bin)
axes[0, 1].hist(x2, bins=n_bin)
axes[0, 2].hist(x3, bins=n_bin)
axes[0, 3].hist(x4, bins=n_bin)
axes[1, 0].hist(x5, bins=n_bin)
axes[1, 1].hist(x6, bins=n_bin)
axes[1, 2].hist(x7, bins=n_bin)
axes[1, 3].hist(x8, bins=n_bin)
axes[2, 0].hist(x9, bins=n_bin)
axes[2, 1].hist(x10, bins=n_bin)
axes[2, 2].hist(x11, bins=n_bin)
axes[2, 3].hist(x12, bins=n_bin)
axes[3, 0].hist(x13, bins=n_bin)
axes[3, 1].hist(x14, bins=n_bin)
axes[3, 2].hist(x15, bins=n_bin)
axes[3, 3].hist(x16, bins=n_bin)
```

### Problem 03 (30 Points)


1. (10 points) Generate independently and uniformly at random `number_of_points = 100` points within a unit square $[0, 1]^2$.

(Hint, a point $(x, y)$ is random within the unit square if both of its coordinates $x$ and $y$ are uniformly and randomly generated from the interval $[0, 1]$.)



2. (10 points) Perform a scattered plot of these points.  



3. Print: the min, max, average, std:

(5 points) calculated over $x$-coordinates,    

(5 points) calculated over $y$-coordinates.   

```python
number_of_points=100
#x and y are uniformly and randomly generated
x = random.uniform(0,1**2,number_of_points)
y = random.uniform(0,1**2,number_of_points)
plt.plot(x,y, 'o')
plt.title('Scatter plot of x and y', fontsize=15, fontweight='bold')
plt.xlabel('randomly generated x')
plt.ylabel('randomly generated y')
```

```python
print("min of x is", min(x))
print("max of x is", max(x))
print("average of x is", np.mean(x))
print("standard deviation of x is", np.std(x))
print("====================================================")
print("min of y is", min(y))
print("max of y is", max(y))
print("average of y is", np.mean(y))
print("standard deviation of y is", np.std(y))
```
